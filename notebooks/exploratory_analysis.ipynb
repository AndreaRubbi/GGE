# Exploratory Data Analysis on Gene Expression Datasets

This notebook provides an exploratory analysis of gene expression datasets. It includes visualizations and statistical summaries to understand the data better and prepare for further evaluations.

```python
# Import necessary libraries
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')

# Load gene expression data
adata = sc.read('path_to_your_data.h5ad')  # Update with your data path

# Display basic information about the dataset
print(adata)

# Visualize the distribution of gene expression levels
plt.figure(figsize=(10, 6))
sc.pl.violin(adata, keys='gene_name', groupby='cell_type', jitter=0.4)
plt.title('Gene Expression Distribution by Cell Type')
plt.show()

# Perform PCA for dimensionality reduction
sc.tl.pca(adata)
sc.pl.pca(adata, color='cell_type')

# Correlation heatmap of the top expressed genes
top_genes = adata.var_names[adata.X.mean(axis=0).argsort()[-20:]]
corr = adata[:, top_genes].to_df().corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Top Expressed Genes')
plt.show()

# Save the processed data for further analysis
adata.write('processed_data.h5ad')  # Update with your desired output path
```
